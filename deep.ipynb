```python
%%writefile app.py
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr

# Check for GPU availability
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32
).to(device)

def generate_image(prompt, negative_prompt="", steps=50, guidance_scale=7.5):
    """
    Generate image from text prompt using Stable Diffusion
    """
    try:
        with torch.autocast(device):
            image = pipe(
                prompt,
                negative_prompt=negative_prompt,
                num_inference_steps=int(steps),
                guidance_scale=guidance_scale
            ).images[0]

        return image

    except Exception as e:
        print(f"Error: {e}")
        return None

# Create Gradio interface
interface = gr.Interface(
    fn=generate_image,
    inputs=[
        gr.Textbox(label="Enter your prompt", lines=3),
        gr.Textbox(label="Negative prompt (optional)", lines=2),
        gr.Slider(10, 100, value=50, label="Inference Steps"),
        gr.Slider(1.0, 15.0, value=7.5, label="Guidance Scale")
    ],
    outputs=gr.Image(label="Generated Image"),
    title="Text-to-Image Generator with Stable Diffusion",
    description="Generate images from text prompts using Stable Diffusion v1-5",
    examples=[
        ["A futuristic cityscape at sunset, digital art"],
        ["A cute corgi puppy wearing a superhero costume, cartoon style"],
        ["A surreal landscape with floating islands and waterfalls"]
    ]
)

# Run the interface
if __name__ == "__main__":
    interface.launch(debug=True, share=True)
```
